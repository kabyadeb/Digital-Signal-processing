### The impulse response of a discrete-time LTI system is h(n)={u(n)−u(n−5)}. Determine the output of the system for the input x[n]=u(n), using the convolution sum.


h(n)={u(n)−u(n−5)

x[n]=u(n) where n=0,1,2,3,4,...,10

so , h(n)=x[n]-u(n-5)


```python
import numpy as np
import matplotlib.pyplot as plt

def unit_step(n):
    return np.where(n>=0,1,0)

n =np.arange(0,11)
x1=unit_step(n)
h=x1-unit_step(n-5)
y=np.convolve(x1,h)

plt.figure(figsize=(12,8))
plt.subplot(3,1,1)
plt.stem(n,x1)
plt.title('Input impulse signal')
plt.xlabel('n')
plt.ylabel('x[n]')
#plt.show()
#for H()
plt.subplot(3,1,2)
plt.stem(n,h)
plt.title('impulse response h[n]')
plt.xlabel('n')
plt.ylabel('h[n]')
#plt.show()
# output 
plt.subplot(3,1,3)
plt.stem(np.arange(0, len(y)), y)
plt.title('Y[n]=x[n]*h[n]')
plt.xlabel('n[n]')
plt.ylabel('y[n]')
plt.show()
```
