```python
import matplotlib.pyplot as plt
import numpy as np
# given 200Hz sampling rate
# and 50Hz analog frequency 2*pi*f
f_analog=50;
time_analog=.2
# time_analog=1/(f_analog);
print("Analog time period: ",time_analog)
fs1=200;
fs2=75;
#time period of the digital signal
n1=np.arange(0,time_analog,1/fs1)
n2=np.arange(0,time_analog,1/fs2)

x1=np.cos(2*np.pi*f_analog*n1 )
x2=np.cos(2*np.pi*f_analog*n2)

plt.Figure(figsize=(10,6))
plt.subplot(2,1,1)
#plt.plot(n1,x1)
plt.stem(n1,x1)
plt.title("No aliasing at 200Hz signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
plt.subplot(2,1,2)
plt.stem(n2,x2)
#plt.plot(n2,x2)
plt.title("Alising at 75 Hz signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid()
plt.show()

```

```python
import numpy as np 
import matplotlib.pyplot as plt

t_max=.005
a1,a2,a3=3,5,10
f1,f2,f3=100,200,300
t = np.linspace(0,t_max,1000)
pii=np.pi
x1=a1*np.cos(2*pii*f1*t)+a2*np.sin(2*pii*f2*t)+a3*np.cos(2*pii*f3*t)

plt.figure(figsize=(14,8))
plt.subplot(4,1,1)
plt.plot(t,x1)
plt.title("analog signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid()
plt.show()

high_time=np.linspace(0,t_max,int(16000*t_max))
x2=a1*np.cos(2*pii*f1*high_time)+a2*np.sin(2*pii*f2*high_time)+a3*np.cos(2*pii*f3*high_time)
plt.subplot(4,1,2)
plt.stem(high_time,x2)
plt.title("high sampling rate")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```
