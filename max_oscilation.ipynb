```python


import numpy as np
import matplotlib.pyplot as plt 

n = np.arange(0,10,1)
print (n)
omega1 =.5*np.pi
omega2 = np.pi

x1 = np.cos(omega1*n)
x2 = np.cos(omega2*n)
plt.Figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.stem(n,x1)
plt.title('x1[n] = sin(0.5*pi*n)')
plt.xlabel('n')
plt.ylabel('x1[n]')
plt.grid()
plt.subplot(1,2,2)
plt.stem(n,x2)
plt.title('x2[n] = sin(pi*n)')
plt.xlabel('n')
plt.ylabel('x2[n]')
plt.grid()

plt.show()
```
