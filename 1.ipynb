## 1 Generating elementary signals like Unit Step, Ramp, Exponential, Sine, and Cosine sequences.


### Sine wave


````python
import numpy as np
import matplotlib.pyplot as plt

fs =1000
f=5
t= np.arange(0,1,1/fs)
x = np.sin(2*np.pi*f*t)
plt.plot(t, x)
plt.title('Sine Wave')
plt.grid()
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
````

### Unite step


````python
import numpy as np
import matplotlib.pyplot as plt 

n=np.arange(-10,11,1)
u_n=np.where(n>=0,1,0)
print(n)
print(u_n)
plt.scatter(n,u_n,label="unit step sequence",)
plt.stem(n,u_n,label="unit step sequence",)
plt.xlabel('n',color='blue')
plt.ylabel('u[n]',color='red')
plt.legend()
plt.grid()
plt.show()
````

### ramp sequence


````python
import numpy as np
import matplotlib.pyplot as plt

n=np.arange(-10,11,.5)
u_n=np.where(n<0,0,n)
# print(n)
# print(u_n)

plt.scatter(n,u_n,label="Unit ramp sequence")
plt.stem(n,u_n)
plt.xlabel('n',color='blue')
plt.ylabel('u[n]',color='red')
plt.title('Unit Ramp Sequence')
plt.legend()
plt.show()
````

### Exponential


````python
import numpy as np
import matplotlib.pyplot as plt

A=100
b=.5
t=np.linspace(-10,10,50)
y=np.exp(b*t)
plt.plot(t,y)
plt.title('Exponential Function')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.legend()
plt.scatter(t,y)
plt.show()
````

### sine wave, cos wave


````python
import numpy as np
import matplotlib.pylab as plt

x=np.linspace(0,2*np.pi,100)
y=np.sin(x) #
yy=np.cos(x) #
plt.plot(x,yy,label='cosine')
plt.plot(x,y,label='sine',color='red')
plt.title('Sine and Cosine Waves')
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

f_original =10
T =1
cont_time=np.linspace(0,T,10000)
High_freq = 50
High_freq_time =np.linspace(0,T,int(High_freq*T))
Low_freq = 15
Low_freq_time =np.linspace(0,T,int(Low_freq*T))

Continious_signal =np.sin(2*np.pi*f_original*cont_time)
High_freq_signal=np.sin(2*np.pi*f_original*High_freq_time)
Low_freq_signal =np.sin(2*np.pi*f_original*Low_freq_time)
plt.figure(figsize=(10,6))
plt.subplot(2,1,1)
plt.plot(cont_time,Continious_signal,label='Continious Signal',color='red')

plt.stem(High_freq_time,High_freq_signal,label='High Frequency Signal')
# plt.stem(High_freq_time,High_freq_signal)
plt.title('High Frequency Signal')
plt.xlabel('Time')
plt.legend()
plt.ylabel('Amplitude')

plt.subplot(2,1,2)
plt.plot(cont_time,Continious_signal,label='Continious Signal',color='red')
plt.stem(Low_freq_time,Low_freq_signal,label='Low Frequency Signal')
plt.title('Low Frequency Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
````
